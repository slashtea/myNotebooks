### Here we are going to answer these questions.
* What is the average speed of a bus being crowded(3), not crowded (2), number of people in it(1)
* What is the average speed of a bus by weather(1: raining, 2: sunny)
* What is the average speed of car_or_bus column 

```python
rating_speed = tracks.groupby('rating')['speed'].mean().sort_values(ascending=False)
rating_speed
```

```python
sns.set()
sns.barplot(x=rating_speed.index, y=rating_speed.get_values())
```

```python
tracks['rating_weather'].value_counts()
```

### Crowded buses are the fastest, that's a surprise !

```python
weather_speed = tracks.groupby('rating_weather')['speed'].mean().sort_values(ascending=False)
weather_speed
```

```python
sns.barplot(x=weather_speed.index, y=weather_speed.get_values())
```

#### What does 0 value stand for ?...dataset description says only 1 and 2. Is it an error ?

* **Anyways, result makes sense, 1(rainy) has a smaller mean speed than 2(sunny)

```python
car_bus_speed = tracks.groupby('car_or_bus')['speed'].mean().sort_values(ascending=False)
car_bus_speed
```

```python
sns.barplot(x=car_bus_speed.index, y=car_bus_speed.get_values())
```

#### Of course cars(1) are faster than bus(2)
